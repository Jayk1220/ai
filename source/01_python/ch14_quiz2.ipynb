```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
#파일 읽기
with open('Data/ch14_맞춤법검사후.txt','r',encoding='utf-8') as f:
    text = f.read()
```

```python
text = input('글 입력 :')
```

```python
#크롬 켜서 papago 접속
driver = webdriver.Chrome()
time.sleep(1)
driver.get('https://papago.naver.com/')
time.sleep(1)

#텍스트 입력 후 번역 버튼 누르기
text_area = driver.find_element(By.ID,"txtSource")
text_area.send_keys(text) # 입력
button = driver.find_element(By.ID,'btnTranslate')
button.click()
time.sleep(1)

#번역 텍스트 가져오기
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source,'html.parser')
return_value = soup.select_one('div#txtTarget').text
# print(return_value)
time.sleep(1)

#브라우저 닫기
driver.close()

#txt파일로 담기
with open('Data/ch14_자동화영어번역본.txt', 'w', encoding="utf-8")as f:
    f.write(return_value)
```
