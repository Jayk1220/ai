## 1.3 그래프 영역 나누기
### 1) subplot()함수로 서브플롯 추가

```python
import numpy as np
x = np.linspace(0,9.9,100) 

x.shape
```

```python
plt.figure(figsize=(5,5))
plt.subplot(2,1,1) #2행 1열로 subplot에서 1번째 subplot은 다음과 같이 그린다
plt.plot(x,np.sin(x))
plt.subplot(2,2,3) #2행 2열 subplot 중 3번째 subplot
plt.plot(x,np.cos(x))
plt.subplot(224) #2행2열 subplot 중 4번째 subplot
plt.scatter(x,np.sin(x)*np.cos(x))
plt.show()
```

```python

plt.subplot(2,2,1) 
plt.plot(x,np.sin(x))

plt.subplot(2,2,2) 
plt.plot(x,np.cos(x))

plt.subplot(2,2,3) 
plt.plot(x,np.tan(x))

plt.subplot(224) 
plt.scatter(x,np.sin(x)*np.cos(x))
plt.show()
```

### 2) subplots()함수로 서브플롯 추가
 - plot 함수에 색상이나 스타일 조정 (교안 14p) https://wi k idocs.net/92085

```python
#2행 2열 서브플롯
fig, axes = plt.subplots(nrows = 2, ncols=2, figsize=(7,4))
axes[0,0].plot(x,np.sin(x), c='r' ,linewidth=10)
axes[0,1].plot(x,np.cos(x), 'c--') #linestyle='--')
axes[1,0].plot(x,np.tanh(x), c='b') #색은 rgb값도 가능
axes[1,0].plot([0,1,2,3],[-1,-0.5,0,0.5],'ko:') #ko ==>black, marker='o',linestyle=':'
axes[1,1].plot(x, np.sin(x)*np.cos(x))
plt.show()
```

```python
#2행 2열 서브플롯 (for문)
func = [np.sin, np.cos, np.tanh, lambda  x : np.sin(x)*np.cos(x)]
fmt = ['r','c--','#00AA00','r-.']
fig,axes = plt.subplots(2,2, figsize=(7,4))
list(enumerate(axes.flat))
for i, ax in enumerate(axes.flat):
    ax.plot(x, func[i](x),fmt[1])
```
