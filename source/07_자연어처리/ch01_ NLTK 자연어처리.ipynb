````python
import nltk
````

````python
#말뭉치 다운로드 받을 폴더
nltk.download()
````

````python
#말뭉치 리스트

nltk.corpus.gutenberg.fileids()
````

````python
# 오스틴 소석 엠마 내용 데이터 셋
emma = nltk.corpus.gutenberg.raw('austen-emma.txt')
print(emma[:30])
````

````python
#sent_tokenize() :문장단위로 쪼갠 lsit 반환
from nltk.tokenize import sent_tokenize
sent_tokens = sent_tokenize(emma)
print('문장수 :', len(sent_tokens)) 
print('%r'%sent_tokens[0])
````

````python
# word_tokenize(): 단어 단위로 쪼갠 list 반환
from nltk.tokenize import word_tokenize
print(word_tokenize(sent_tokens[0]))
````

````python
# RegexpTokenizer 클래스 : 토큰화 할 때, 정규표현식
from nltk.tokenize import RegexpTokenizer
ret = RegexpTokenizer('\d+')
digits = ret.tokenize(emma)
print(digits)
ret1 = RegexpTokenizer('\w+') #[a-zA-Z0-9-] : \w
words = ret1.tokenize(sent_tokens[0])
print(words)
````

# 형태소 분석(의미가 있는 가장 작은 단위) 분석
cf. 자연어 처리의 기본은 형태소 분석과 품사 태깅 - 어간 추출(stemming), 품사 태깅(PosTagging)

````python
words = ['sending', 'cooking', 'files','lives','crying','dying']
#어간추출 1 : PosterStemmer
from nltk.stem import PorterStemmer, LancasterStemmer, RegexpStemmer
pst = PorterStemmer()
# pst.stem(words[0], words[1])

[pst.stem(word)for word in words] 
````

````python
#어간추출 2: LanscasterStemmer
lst = LancasterStemmer()
[lst.stem(word)for word in words] 
````

````python
#어간추출3 : RegexpStemmer
rst = RegexpStemmer('ing')
[rst.stem(word)for word in words] 
````

# 품사태깅

````python
print('품사 태깅할 내용 : ', word_tokenize(sent_tokens[0]))
from nltk.tag import pos_tag
tagged_list = pos_tag(word_tokenize(sent_tokens[0]))
print('품사태깅 결과: ', tagged_list)
````

````python
#명사만 추출 : NN NNS ,NNP< NNPS, PRP
nouns_list = []

for word, tag in tagged_list:
#     if tag in ['NN','NNS','NNP','NNPS','PRP']:
    if tag.find('NN') != -1:
        nouns_list.append(word)
print(nouns_list)
````

````python
nouns_list = [word for word, tag in tagged_list if tag.find('NN') != -1]
print(nouns_list)
````

1. 특수문자가 들어가지 않는 3글자 이상의 단어만 추출해서 품사태깅 (RegexpTokenizer)


````python
emma = nltk.corpus.gutenberg.raw('austen-emma.txt')
emma = emma.lower()
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
ret1 = RegexpTokenizer('[a-zA-Z]+') #[a-zA-Z0-9-] : \w
words = ret1.tokenize(emma)

wordsover3 = [x for x in words if len(x) >= 3]
tagged_list = pos_tag(wordsover3)
tagged_list
````
