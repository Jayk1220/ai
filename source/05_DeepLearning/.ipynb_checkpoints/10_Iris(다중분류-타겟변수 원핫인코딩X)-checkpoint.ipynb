```python
data = sns.load_dataset('iris')
data.info()
```

```python
#인코딩
#라벨
le = LabelEncoder()
label_d = le.fit_transform(data['species'])
#원핫
one_hot_d = to_categorical(label_d)


X = data.iloc[:,:-1].values
```

```python
#train test separation
X_train, X_test, Y_train, Y_test = train_test_split(X, one_hot_d,
                                                    train_size=0.8,
                                                    stratify = one_hot_d,
                                                    random_state = 15
)

X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```
