```python
import numpy as np
from tensorflow.keras.models import Sequential # 모델 객체 생성
from tensorflow.keras.layers import Input, Dense #모델 layer 쌓을때 필요
import matplotlib.pyplot as plt
```

```python
#data set
x_data = np.array([[0,0],
                   [0,1],
                   [1,0],
                   [1,1]])
y_data = np.array([[0],[1],[1],[0]])

#2. 모델구성
model = Sequential()
# model.add(Input(shape=2))
model.add(Dense(units = 10, #출력 갯수
                input_dim = 2, #input_shape = (2,) #입력갯수
                activation = 'relu')) #활성화 함수
model.add(Dense(units=20, activation = 'relu')) #은닉층에서는 input_shape는 x

model.add(Dense(10, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
print(model.summary())

#3. 모델 학습과정 설정
    #loss : mae(회귀분석), binary_crossentropy(이진분류=로지스틱회귀분석),
    #categorical_corssentropy(다중분류 - 원핫인코딩을 한 경우)
    #sparse_categorical_crossentropy(다중분류-원핫인코딩을 안 한 경우)
model.compile(loss='binary_crossentropy', #손실함수 (loss function)
              optimizer='adam',
              metrics= ['binary_accuracy']) #이진분류일 경우
```

```python
#4 학습 - 1번만 실행
hist = model.fit(x_data, y_data, epochs=300, verbose=1)
```

```python
#5. 학습과정 살펴보기
hist.history.keys()
```

```python
fig, loss_ax = plt.subplots(figsize=(10,5))
loss_ax.plot(hist.history['loss'],'b',label='loss')
acc_ax = loss_ax.twinx()#loss_ax와 x축을 공유하는 acc_ax
acc_ax.plot(hist.history['binary_accuracy'],'r',label='binary_accuracy')
loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('binary_accuracy')
loss_ax.legend(loc='center right')
acc_ax.legend(loc='center left')
plt.show()
```

```python
#모델 평가하기
score = model.evaluate(x_data, y_data) #loss & metrics (binary accuracy)
print('loss : ', score[0])
print('accuracy :', score[1])
```
